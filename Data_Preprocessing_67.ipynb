```python
#จากข้อมูลที่ให้มา มีค่าว่างสูญหายมากน้อยแค่ไหน? และมีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?

import matplotlib.pyplot as plt
# ตรวจสอบค่าว่าง
missing_values = titanic.isnull().sum()
print("Missing Values:\n", missing_values)

# ตรวจสอบค่าผิดปกติ (ตัวอย่าง)
plt.figure(figsize=(20, 5))
plt.boxplot(titanic['age'].dropna(), vert=False)
plt.title('Boxplot of Age')
plt.xlabel('age')
plt.show()

# ตรวจสอบข้อมูลรบกวน (ตัวอย่าง)
plt.figure(figsize=(8, 6))
plt.hist(titanic['fare'], bins=20)
plt.title('Histogram of Fare')
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.show()
```

* (10 points) จากข้อมูลที่ให้มา มีค่าว่างสูญหายมากน้อยแค่ไหน? และมีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?

```python
# จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

# จัดการกับค่าว่างใน age ด้วยการแทนที่ด้วยค่าเฉลี่ย
titanic['age'].fillna(titanic['age'].mean(), inplace=True)

# จัดการกับค่าว่างใน embarked ด้วยการแทนที่ด้วยค่าที่พบบ่อยที่สุด
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

# จัดการกับค่าว่างใน cabin ด้วยการลบทั้งคอลัมน์ เนื่องจากมีค่าว่างมากเกินไป
titanic.drop('cabin', axis=1, inplace=True)

# ตรวจสอบค่าว่างอีกครั้ง
missing_values_after = titanic.isnull().sum()
print("Missing Values After Handling:\n", missing_values_after)

# อธิบายวิธีที่เลือกใช้
explanation = """
วิธีที่เลือกใช้ในการจัดการกับค่าว่างในข้อมูล

•Age : เลือกใช้การแทนที่ด้วยค่าเฉลี่ย เนื่องจากเป็นตัวแปรเชิงปริมาณ การแทนที่ด้วยค่าเฉลี่ยจะไม่ทำให้การกระจายของข้อมูลเปลี่ยนแปลงไปมากนัก และช่วยรักษาลักษณะของข้อมูลให้คงที่

•Embarked : เลือกใช้การแทนที่ด้วยค่าที่พบบ่อยที่สุด (Mode) เนื่องจากเป็นตัวแปรเชิงหมวดหมู่ การแทนที่ด้วยค่าที่พบบ่อยที่สุดจะไม่ทำให้เกิดความเบี่ยงเบนในข้อมูลมากนัก และยังคงความสมบูรณ์ของข้อมูล

•Cabin : เลือกใช้การลบทั้งคอลัมน์ เนื่องจากคอลัมน์นี้มีค่าว่างจำนวนมาก การแทนที่ด้วยค่าอื่นอาจทำให้เกิดความคลาดเคลื่อนหรือข้อมูลที่ไม่สมบูรณ์ในลักษณะเดิมได้ ดังนั้น การลบทั้งคอลัมน์จะช่วยลดความเสี่ยงนี้
"""
print(explanation)
```

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
# prompt: แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข
# i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked
# ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1
# iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard)
# iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็น

import pandas as pd
# i. ใช้ one-hot encoding ในการแปลงตัวแปร Embarked
titanic = pd.get_dummies(titanic, columns=['embarked'])

# ii. แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1
titanic['sex'] = titanic['sex'].map({'male': 0, 'female': 1})

# iii. สร้างตัวแปร FamilySize
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']

# iv. สร้างตัวแปร IsAlone
titanic['IsAlone'] = titanic['FamilySize'].apply(lambda x: 1 if x == 0 else 0)

# แสดงข้อมูลหลังจากการแปลง
titanic
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
# prompt: ทำการ Normalize ข้อมูล
# i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling
# ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1
# iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

from sklearn.preprocessing import MinMaxScaler

# i. Min-Max Scaling for Fare and Age
scaler = MinMaxScaler()
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])

# ii. Create Fare_Per_Person
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)

# iii. Create AgeGroup
def categorize_age(age):
  if 0 <= age < 12:
    return '[0, 12)'
  elif 12 <= age < 18:
    return '[12, 18)'
  elif 18 <= age < 50:
    return '[18, 50)'
  else:
    return '[50, 100)'

titanic['AgeGroup'] = titanic['age'].apply(categorize_age)

titanic
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling \
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ตั้งค่าธีมสำหรับ Seaborn เพื่อปรับปรุงความสวยงาม
sns.set_theme(style="whitegrid")

# i. survived และ pclass
survival_by_pclass = titanic.groupby('pclass')['survived'].mean()
print(survival_by_pclass)

plt.figure(figsize=(8, 6))
survival_by_pclass.plot(kind='bar', color=sns.color_palette("muted"))
plt.title('Survival Rate by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=14)
plt.ylabel('Survival Rate', fontsize=14)
plt.xticks(rotation=0)
plt.show()

# ii. survived และ sex
survival_by_sex = titanic.groupby('sex')['survived'].mean()
print(survival_by_sex)

plt.figure(figsize=(8, 6))
survival_by_sex.plot(kind='bar', color=sns.color_palette("muted", 2))
plt.title('Survival Rate by Sex', fontsize=16)
plt.xlabel('Sex (0=Male, 1=Female)', fontsize=14)
plt.ylabel('Survival Rate', fontsize=14)
plt.xticks(rotation=0)
plt.show()

# iii. survived และ age
plt.figure(figsize=(40, 5))
plt.scatter(titanic['age'], titanic['survived'], c=titanic['survived'], cmap='viridis', alpha=0.7)
plt.title('Survival vs Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Survived (1=Yes, 0=No)', fontsize=14)
plt.colorbar(label='Survival')
plt.show()

# iv. จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare
family_survival = titanic.groupby('FamilySize')['survived'].sum()
fare_avg = titanic.groupby('FamilySize')['fare'].mean()

plt.figure(figsize=(8, 6))
plt.scatter(family_survival.index, family_survival.values,
            s=fare_avg*10, c=fare_avg, cmap='coolwarm', alpha=0.7, edgecolors='w')
plt.title('Number of Family Members Survived vs Average Fare', fontsize=16)
plt.xlabel('Family Size', fontsize=14)
plt.ylabel('Number of Survived Family Members', fontsize=14)
plt.colorbar(label='Average Fare')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# สร้าง Pivot Table เพื่อดูอัตราการรอดชีวิตโดยแบ่งตามเพศและชั้นโดยสาร
pivot_table = titanic.pivot_table(values='survived', index='sex', columns='pclass', aggfunc='mean')

# แสดง Pivot Table พร้อมการตกแต่งตาราง
print("Pivot Table - Survival Rate by Sex and Passenger Class")
display(pivot_table.style.background_gradient(cmap='YlGnBu').format("{:.2%}"))

# สร้าง Heatmap เพื่อแสดงผล Pivot Table ให้ดูง่ายขึ้น
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, fmt=".2%", cmap='YlGnBu', linewidths=.5)
plt.title('Survival Rate by Sex and Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=14)
plt.ylabel('Sex', fontsize=14)
plt.show()

# สร้าง Pivot Table เพื่อดูอายุเฉลี่ยของผู้โดยสารที่รอดชีวิตและไม่รอดชีวิต โดยแบ่งตามเพศ
pivot_table_age = titanic.pivot_table(values='age', index='survived', columns='sex', aggfunc='mean')

# แสดง Pivot Table พร้อมการตกแต่งตาราง
print("Pivot Table - Average Age of Survivors and Non-Survivors by Sex")
display(pivot_table_age.style.background_gradient(cmap='coolwarm').format("{:.2f}"))

# สร้าง Bar Chart เพื่อแสดงผล Pivot Table ให้ดูง่ายขึ้น
pivot_table_age.plot(kind='bar', figsize=(8, 6), color=sns.color_palette("Set2"))
plt.title('Average Age of Survivors and Non-Survivors by Sex', fontsize=16)
plt.xlabel('Survived (0=No, 1=Yes)', fontsize=14)
plt.ylabel('Average Age', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Sex', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# สรุปผลที่ได้จากตาราง
summary = """
จากตารางสรุปข้อมูลและแผนภูมิเราสามารถสรุปได้ว่า

  ผู้หญิงมีอัตราการรอดชีวิตสูงกว่าผู้ชายอย่างมีนัยสำคัญ โดยเฉพาะผู้หญิงในชั้นโดยสารที่ 1 และ 2 มีอัตราการรอดชีวิตสูงกว่าผู้ชาย
ในชั้นเดียวกันมาก ผู้โดยสารในชั้นโดยสารที่ 1 มีอัตราการรอดชีวิตสูงกว่าผู้โดยสารในชั้นอื่น นี่แสดงให้เห็นถึงความเหลื่อมล้ำในสังคม
ในยุคนั้น อายุเฉลี่ยของผู้รอดชีวิตต่ำกว่าอายุเฉลี่ยของผู้ที่ไม่รอดชีวิต นี่อาจบ่งชี้ว่าเด็กและคนหนุ่มสาวมีโอกาสรอดชีวิตมากกว่าผู้สูงอายุ
"""
print(summary)
```
